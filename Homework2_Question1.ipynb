```python
from selenium.webdriver import Keys
import re
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from lxml import etree
import pandas as pd
from lxml import etree

#创建webdriver对象,启动谷歌浏览器
driver = webdriver.Chrome('../chromedriver.exe')

#打开网址
driver.get('https://www.ireader.com/index.php?ca=channel.index&pca=search.index')

#等待变量
wait = WebDriverWait(driver,10)

#定位搜索框
search_btn = driver.find_element(By.CSS_SELECTOR, '#search_form > input[type=text]')

#在搜索框输入
search_btn.send_keys('儿童')
search_btn.send_keys(Keys.RETURN)
```

```python
name = []
new_writer = []
i=1
while i<=1667:
    dom = etree.HTML(driver.page_source, etree.HTMLParser())

    rqq = driver.page_source
    name.extend(dom.xpath('/html/body/div[2]/div[1]/ul/li[position()]/div/div/a/h3/text()'))
    writer = dom.xpath('/html/body/div[2]/div[1]/ul/li[position()]/div/div/p[1]/text()')
    new_writer.extend([re.findall('[\u4e00-\u9fa5：·\w]+', i)[0][3:] for index, i in enumerate(writer) if index % 3 == 0])

    if '下一页' not in dom.xpath('/html/body/div[2]/div[1]/div[2]/a/text()'):   # 判定是否还有“后页”按钮农田
        print('爬取结束')
        break

    driver.find_element_by_link_text('下一页').click()

    print('第['+str(i)+']页爬取成功')
    i+=1
```

```python
pd.DataFrame({'书名':name,'作者':new_writer})
```
